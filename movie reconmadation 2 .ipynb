```python

def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert3) 
movies.head()
```

```python
movies['cast'].head()
```

```python
def fetch_director(text): 
    L = [] 
    for i in ast.literal_eval(text):
        if i ['job'] =='Director':
            L.append(i['name'])
    return L         
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
def space(sp):
    L1 = []
    for i in sp:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(space) 
movies['crew'] = movies['crew'].apply(space) 
movies['genres'] = movies['genres'].apply(space) 
movies['keywords'] = movies['keywords'].apply(space) 
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split()) 
```

```python
movies.head(3)
```

```python
movies['marges'] =  movies['cast'] +movies['crew'] + movies['keywords'] + movies['genres'] +movies['overview']
```

```python
new = movies.drop(columns=['genres','keywords','overview','cast','crew']) 
```

```python
new['marges'] = new['marges'].apply(lambda x:"".join(x))
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
coun = CountVectorizer(max_features=500,stop_words='english') 
```

```python
vector = coun.fit_transform(new['marges']).toarray()
```

```python
vector.shape 
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector) 
```
